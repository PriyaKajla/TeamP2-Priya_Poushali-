```
import pandas as pd
data= pd.read_csv("titanic.csv")
```

```
data.head(5)
```

```
data.isnull()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
pal = {'male':"skyblue", 'female':"pink"}
plt.subplots(figsize=(10,6))
ax = sns.barplot(x = "Sex",y = "Survived",data=data,palette = pal,linewidth=2 )
plt.title("Survived/Non-Survived Passenger Gender Distribution", fontsize = 25)
plt.ylabel("% of passenger survived", fontsize = 15)
plt.xlabel("Sex",fontsize = 15)
plt.show()

```

~74% female passenger survived while only ~19% male passenger survived.

```
pal = {1:"red", 0:"blue"}
sns.set(style="darkgrid")
plt.subplots(figsize = (10,8))
ax = sns.countplot(x = "Sex",hue="Survived",data = data,linewidth=1,palette = pal)

plt.title("Passenger Gender Distribution - Survived vs Not-survived", fontsize = 20)
plt.xlabel("Sex", fontsize = 10);
plt.ylabel("# of Passenger Survived", fontsize = 10)

leg = ax.get_legend()
leg.set_title("Survived")
legs = leg.texts
legs[0].set_text("No")
legs[1].set_text("Yes")
plt.show()
```

It shows that among all the females ~ 230 survived and ~ 70 did not survive. While among male passengers ~110 survived and ~470 did not survive.


```
fig, ax= plt.subplots(figsize=(10,6))
sns.violinplot(ax = ax, y=data["Pclass"] )
plt.show()
```

Maximum number of passengers belong to lower socio-economic class. 

```
fig = plt.figure(figsize=(10,8),)
ax=sns.kdeplot(data.loc[(data['Survived'] == 0),'Fare'] , color='blue',shade=True,label='not survived')
ax=sns.kdeplot(data.loc[(data['Survived'] == 1),'Fare'] , color='g',shade=True, label='survived')
plt.title('Fare Distribution Survived vs Non Survived', fontsize = 20)
plt.ylabel("Frequency of Passenger Survived", fontsize = 10)
plt.xlabel("Fare", fontsize = 15)
plt.show()
```

The spike in the plot under 100 dollar represents that a lot of passengers who bought the ticket within that range did not survive. 

```
plt.subplots(figsize = (10,6))
sns.barplot(x = "Pclass",y = "Survived",data=data,linewidth=1)
plt.title("Passenger Class Distribution - Survived vs Non-Survived", fontsize = 20)
plt.xlabel("Socio-Economic class", fontsize = 10);
plt.ylabel("% of Passenger Survived", fontsize = 10);
labels = ['Upper', 'Middle', 'Lower']
val = [0,1,2] 
plt.xticks(val, labels)
plt.show()
```


    ~ 63% first class passenger survived titanic tragedy, while
    ~ 48% second class and
    ~ only 24% third class passenger survived.


```
fig, (axis1) = plt.subplots(1,figsize=(10,8))

sns.violinplot(x = 'Pclass', y = 'Age', hue = 'Survived', data = data, split = True, ax = axis1)
axis2.set_title('Pclass vs Age Survival Comparison')
plt.show()
```

First Class: Max passengers survived  -  ~35-38 yrs old, max passengers who died - ~45yrs old.

Second Class: Max passengers survived  -  ~30 yrs old, max passengers who died - ~26yrs old.

Third Class: Max passengers survived  -  ~25 yrs old, max passengers who died - ~22yrs old.


```
data = data[data.Fare < 500]
sns.factorplot(x = "Parents/Children Aboard", y = "Survived", data = data,kind = "point",size = 8)
plt.title("Factorplot of Parents/Children survived", fontsize = 20)
plt.subplots_adjust(top=0.85)

plt.show()
```

Passenger who traveled in big groups with parents/children had less survival rate than other passengers.

```
sns.factorplot(x =  "Siblings/Spouses Aboard", y = "Survived", data = data,kind = "point",size = 8)
plt.title('Factorplot of Sibilings/Spouses survived', fontsize = 25)
plt.subplots_adjust(top=0.85)

plt.show()
```

While, passenger who traveled in small groups with sibilings/spouses had more survival rate than other passengers.
